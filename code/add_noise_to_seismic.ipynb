```python
# import necesssary packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the npy file containing the seismic section
input_path = '../data/processed/seismics/'
X_test = np.load(input_path + 'slice1.npy')
plt.imshow(X_test)
plt.set_cmap('gray')
```

```python
# Add a random amount of uniform noise to the image
mu_fact = 0.15
sigma_fact = 0.15
amax_noise = np.max(X_test)
mu = amax_noise * mu_fact
sigma = amax_noise * sigma_fact
noise = np.random.normal(mu, sigma, X_test.shape)
noise[noise<0] = 0
X_test_noisy = X_test + noise
plt.imshow(X_test_noisy)

# save the noisy file in the same location
np.save(input_path + 'slice1_noisy.npy', X_test_noisy)
```
