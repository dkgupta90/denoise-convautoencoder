```python
# importing the needed packages
import numpy as np
import segyio as sg
import matplotlib.pyplot as plt
```

```python
# Reading the seismic data into a numpy array
seismic_fname = '../data/raw/2D_Seismic_Alaska/U26_22.SGY'
f = sg.open(seismic_fname, ignore_geometry = True)

X_test = f.trace
print(X_test.shape, X_test)
# Manually creating the array
ntraces = 4228
nsamples = 3071
X_test = np.empty((nsamples, ntraces))
for trace in range(ntraces):
    X_test[:, trace] = f.trace[trace]
plt.figure(figsize = (15, 30))
sliced_sec = X_test[201:400, 2001:2400]
plt.imshow(sliced_sec)
plt.set_cmap('gray')

# saving the seismic section
write_path = '../data/processed/seismics/'
np.save(write_path + 'slice2.npy', sliced_sec)
```
